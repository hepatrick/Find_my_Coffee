```python
import requests
import pandas as pd
from yelpapi import get_my_key
import json
```

```python
business_id = "e1zDuraqPYNqfEmgXJJAyQ"

#API_KEY = get_my_key

ENDPORINT = "https://api.yelp.com/v3/businesses/search"

HEADERS = {'Authorization': 'bearer %s' % get_my_key}
```

```python
PARAMETERS = {"term": "coffee", "limit": 50, "radius": 5000, "location": "Chicago"}

response = requests.get(url = ENDPORINT, params= PARAMETERS, headers=HEADERS)

business_data = response.json()

business_data
```

```python
good_coffee = business_data["businesses"]
good_coffee_df = pd.DataFrame(good_coffee)
good_coffee_df = good_coffee_df[["name","is_closed","categories","review_count","rating","location"]]
good_coffee_df.head()
```

```python
great_coffee_shop = good_coffee_df.loc[(good_coffee_df["review_count"]>100) & ( good_coffee_df["rating"] >=4.0)]
great_coffee_shop
```

```python
num_cafe = great_coffee_shop["name"].count()
num_cafe
```

```python
zip_list = []

for i in num_cafe:
    temp_zip = great_coffee_shop["location"][i]["zip_code"]
    zip_list.append(temp_zip)
    
zip_list
```
